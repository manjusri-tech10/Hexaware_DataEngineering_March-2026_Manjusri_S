flatMap()
Applies a function to each element
Returns multiple outputs per input

```python
lines = sc.parallelize([
    "python spark hadoop",
    "spark sql pyspark"
])

words = lines.flatMap(lambda line: line.split(" "))
words.collect()
```

```python
lines = sc.parallelize(["hello world", "pyspark rdd"])

result = lines.map(lambda x: x.split(" "))
result.collect()
```

```python
lines = sc.parallelize(["hello world", "pyspark rdd"])

result = lines.flatMap(lambda x: x.split(" "))
result.collect()
```

```python
rdd = sc.parallelize([10, 20, 20, 30, 30, 30, 40])

rdd.distinct().collect()#diff elem
```

```python
rdd1 = sc.parallelize([1, 2, 3])
rdd2 = sc.parallelize([4, 5, 6])

rdd1.union(rdd2).collect()# combine
```

```python
rdd1 = sc.parallelize([1, 2, 3, 4])
rdd2 = sc.parallelize([3, 4, 5, 6])

rdd1.intersection(rdd2).collect()#common elem btwn 2 rdd
```

```python
rdd1 = sc.parallelize([1, 2, 3, 4])
rdd2 = sc.parallelize([3, 4])

rdd1.subtract(rdd2).collect()#sub comm elem
```

```python
rdd = sc.parallelize([
    ("Rahul", 85),
    ("Sneha", 92),
    ("Arjun", 78)
])

rdd.mapValues(lambda x: x + 5).collect()#works op only on values
```

```python
lines = sc.parallelize([
    "python spark python",
    "spark hadoop spark",
    "python hadoop"
])

word_counts = (
    lines
    .flatMap(lambda line: line.split(" "))
    .map(lambda word: (word, 1))
    .reduceByKey(lambda a, b: a + b)
)

word_counts.collect()
```

```python
employees = sc.parallelize([
    ("Rahul", 70000),
    ("Sneha", 60000),
    ("Arjun", 75000),
    ("Priya", 80000),
    ("Karan", 50000)
])
high_salary = employees.filter(lambda x: x[1] > 65000)
high_salary.collect()
```
