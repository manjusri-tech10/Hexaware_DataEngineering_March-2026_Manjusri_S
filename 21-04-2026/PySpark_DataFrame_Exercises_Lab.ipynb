```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, sum, avg, max, min, lit, upper, length, when, to_date
spark = SparkSession.builder.appName("LabPractice").getOrCreate()

df_emp   = spark.read.csv("employees.csv", header=True, inferSchema=True)
df_dept  = spark.read.csv("departments.csv", header=True, inferSchema=True)
df_sales = spark.read.csv("sales (1).csv", header=True, inferSchema=True)
df_json  = spark.read.option("multiLine", True).json("employees_nested.json")

print("All DataFrames loaded successfully!")
```

```python
df_emp.show(5)
```

```python
print("Total employees:", df_emp.count())
```

```python
print("Columns:", df_emp.columns)
```

```python
df_emp.show(3)
```

```python
df_emp.select("name", "salary").show()
```

```python
df_emp.withColumnRenamed("salary", "emp_salary").show()
```

```python
from pyspark.sql.functions import col, to_date

df_emp.filter(col("joining_date") > "2023-01-01").show()
```

```python
df_emp.filter(col("salary") > 65000).show()
```

```python
df_emp.filter(col("department") == "IT").show()
```

```python
df_emp.filter((col("department") == "IT") & (col("salary") > 70000)).show()
```

```python
df_emp.drop("joining_date").show()
```

```python
df_emp.orderBy("salary").show()
```

```python
df_emp.orderBy(col("salary").desc()).show()
```

```python
df_emp.orderBy(col("salary").desc()).limit(3).show()
```

```python
df_emp.select(min("salary").alias("min_salary")).show()
```

```python
df_emp.groupBy("department").count().show()
```

```python
df_emp.groupBy("department").avg("salary").show()
```

```python
df_emp.groupBy("department").sum("salary").show()
```

```python
df_inner = df_emp.join(df_dept, on="department", how="inner")
df_inner.show()
```

```python
df_left = df_emp.join(df_dept, on="department", how="left")
df_left.show()
```

```python
df_emp.join(df_dept, on="department", how="inner") \
    .select("name", "department", "location").show()
```

```python
df_emp.withColumn("bonus", col("salary") * 0.10).show()
```

```python
from pyspark.sql.functions import lit

df_emp.withColumn("company", lit("BotCampus")).show()
```

```python
df_emp.filter(col("department").isNull()).show()
```

```python
df_emp.fillna({"department": "Unknown"}).show()
```

```python
df_emp.dropna().show()
```

```python
from pyspark.sql.functions import upper

df_emp.withColumn("name_upper", upper(col("name"))).show()
```

```python
from pyspark.sql.functions import length

df_emp.withColumn("name_length", length(col("name"))).show()
```

```python
df_emp.withColumn("salary_in_lakhs", col("salary") / 100000).show()
```
