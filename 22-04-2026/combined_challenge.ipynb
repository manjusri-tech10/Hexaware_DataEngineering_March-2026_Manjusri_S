```python
from google.colab import files
import pandas as pd
import json
import io
from collections import Counter

logins      = uploaded['logins (1).txt'].decode('utf-8').strip().split('\n')
employees_df = pd.read_csv(io.BytesIO(uploaded['employees (1).csv']))
sales_df     = pd.read_csv(io.BytesIO(uploaded['sales (2).csv']))
orders_data  = json.load(io.BytesIO(uploaded['orders (1).json']))
orders       = orders_data['orders']

employees_df['salary'] = employees_df['salary'].astype(int)
sales_df['quantity']   = sales_df['quantity'].astype(int)
sales_df['price']      = sales_df['price'].astype(int)
```

```python
print("All login names:")
print(logins)
```

```python
print(f"Total login events: {len(logins)}")
```

```python
print(f"Unique users: {sorted(set(logins))}")
```

```python
login_counts = Counter(logins)
print("Login count per user:")
for user, count in sorted(login_counts.items()):
    print(f"   {user}: {count}")
```

```python
print("Top 3 most active users:")
for user, count in login_counts.most_common(3):
    print(f"   {user}: {count} logins")
```

```python
print("Users with more than 4 logins:")
for user, count in login_counts.items():
    if count > 4:
        print(f"   {user}: {count}")
```

```python
print("Login dictionary:")
print(dict(login_counts))
```

```python
print("CSV: Employees Dataset")
print(f"CSV loaded. Columns: {list(employees_df.columns)}")
```

```python
print(f"Total employees: {len(employees_df)}")
```

```python
it_df = employees_df[employees_df['department'] == 'IT']
print(f"IT Department employees ({len(it_df)}):")
print(it_df[['name','salary','city']].to_string(index=False))
```

```python
high_sal = employees_df[employees_df['salary'] > 75000]
print(f"Salary > ₹75,000 ({len(high_sal)}):")
print(high_sal[['name','department','salary']].to_string(index=False))
```

```python
print(f"Average salary: ₹{employees_df['salary'].mean():,.2f}")
```

```python
idx = employees_df['salary'].idxmax()
print(f"Highest paid: {employees_df.loc[idx,'name']} — ₹{employees_df.loc[idx,'salary']}")
```

```python
idx = employees_df['salary'].idxmin()
print(f"Lowest paid: {employees_df.loc[idx,'name']} — ₹{employees_df.loc[idx,'salary']}")
```

```python
print("Employees per department:")
print(employees_df['department'].value_counts().to_string())
```

```python
print("Average salary per department:")
print(employees_df.groupby('department')['salary'].mean().apply(lambda x: f"₹{x:,.2f}").to_string())
```

```python
print("Employees per city:")
print(employees_df['city'].value_counts().to_string())
```

```python
print("Top 5 highest salaries:")
print(employees_df.nlargest(5,'salary')[['name','salary']].to_string(index=False))
```

```python
hyd_high = employees_df[(employees_df['city']=='Hyderabad') & (employees_df['salary']>70000)]
print(f"Hyderabad & salary greater than ₹70k ({len(hyd_high)}):")
print(hyd_high[['name','salary']].to_string(index=False))
```

```python
print("CSV: Sales Dataset")
sales_df['revenue'] = sales_df['quantity'] * sales_df['price']
print("Revenue per sale:")
print(sales_df[['sale_id','product','quantity','price','revenue']].to_string(index=False))
```

```python
print(f"Total revenue: ₹{sales_df['revenue'].sum():,}")
```

```python
print("Revenue per product:")
prod_rev = sales_df.groupby('product')['revenue'].sum()
for prod, rev in prod_rev.items():
    print(f"   {prod} → ₹{rev:,}")
```

```python
print("Total quantity sold per product:")
print(sales_df.groupby('product')['quantity'].sum().to_string())
```

```python
best = prod_rev.idxmax()
print(f"Best selling product (by revenue): {best} — ₹{prod_rev[best]:,}")
```

```python
emp_rev_s = sales_df.groupby('emp_id')['revenue'].sum()
top_id = emp_rev_s.idxmax()
top_name = employees_df[employees_df['emp_id']==top_id]['name'].values[0]
print(f"Top revenue employee: {top_name} (emp_id={top_id}) — ₹{emp_rev_s[top_id]:,}")
```

```python
print(f"Average sale value: ₹{sales_df['revenue'].mean():,.2f}")
```

```python
print("Products with revenue > ₹1,00,000:")
for prod, rev in prod_rev.items():
    if rev > 100000:
        print(f"   {prod} → ₹{rev:,}")
```

```python
print("JSON: Orders Dataset")
orders_df = pd.DataFrame(orders)
print("JSON loaded successfully.")
```

```python
print("All orders:")
print(orders_df.to_string(index=False))
```

```python
print(f"Total orders: {len(orders_df)}")
```

```python
print(f"Total sales amount: ₹{orders_df['amount'].sum():,}")
```

```python
print("Total spending per customer:")
cust_spend = orders_df.groupby('customer')['amount'].sum()
for c, a in cust_spend.items():
    print(f"   {c}: ₹{a:,}")
```

```python
top_c = cust_spend.idxmax()
print(f"Highest spending customer: {top_c} — ₹{cust_spend[top_c]:,}")
```

```python
print("Total sales per product:")
for p, a in orders_df.groupby('product')['amount'].sum().items():
    print(f"   {p}: ₹{a:,}")
```

```python
hyd_orders = orders_df[orders_df['city']=='Hyderabad']
print(f"Customers from Hyderabad ({len(hyd_orders)}):")
print(hyd_orders[['customer','product','amount']].to_string(index=False))
```

```python
big = orders_df[orders_df['amount'] > 10000]
print(f"Orders with amount > ₹10,000 ({len(big)}):")
print(big[['order_id','customer','product','amount']].to_string(index=False))
```

```python
print("Orders per city:")
print(orders_df['city'].value_counts().to_string())
```

```python
print("FINAL COMBINED CHALLENGE")
merged = sales_df.merge(employees_df[['emp_id','name','department']], on='emp_id', how='left')
print(f"\nTask 1. Joined {len(merged)} records on emp_id")
```

```python
emp_total = merged.groupby('name')['revenue'].sum().sort_values(ascending=False)
print("Total revenue per employee:")
for name, rev in emp_total.items():
    print(f"   {name} → ₹{rev:,}")
```

```python
print("Top 5 employees by revenue:")
for i, (name, rev) in enumerate(emp_total.head(5).items(), 1):
    print(f"   {i}. {name} → ₹{rev:,}")
```

```python
dept_rev = merged.groupby('department')['revenue'].sum().sort_values(ascending=False)
print("Revenue per department:")
for dept, rev in dept_rev.items():
    print(f"{dept}: ₹{rev:,}")
print(f"\nTop department: {dept_rev.idxmax()} — ₹{dept_rev.max():,}")
```

```python
report = merged.groupby(['name','department'])['revenue'].sum().reset_index()
report.columns = ['Employee', 'Department', 'Total Revenue']
report = report.sort_values('Total Revenue', ascending=False)
report.to_csv('final_sales_report.csv', index=False)
```
